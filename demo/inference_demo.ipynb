```python
from mmdet3d.apis import init_model, inference_detector, show_result_meshlab, inference_multi_modality_detector
```

```python
config_file = '../configs/transfusion/transfusion_nusc_pillar_LC.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
# checkpoint_file = '/home/project/TransFusion/checkpoints/transfusion_git/transfusion_LC.pth'
checkpoint_file = '../work_dirs/transfusion_nusc_pillar_LC/latest.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_model(config_file, checkpoint_file, device='cuda:0')
```

```python
# test a single sample

result, data = inference_multi_modality_detector(model, '/home/project/TransFusion/data/output_sample/lidar_top.pcd.bin',
    '/home/project/TransFusion/data/output_sample/CAM_BACK_LEFT.jpg',
    '/home/project/TransFusion/data/output_sample/annotations.json'
)
```

```python
# show the results
out_dir = './'
show_result_meshlab(data, result, out_dir)
```
